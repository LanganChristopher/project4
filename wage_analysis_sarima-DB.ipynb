```python


def clean_tide_max(df):
    df = df.drop(columns= ['Predicted (ft)', 'Preliminary (ft)'], axis=1)
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.set_index('Date')
    #df.index = pd.to_datetime(df.index, format='%Y-%m')
    df = df[df['Verified (ft)'] != '-']
    df['Verified (ft) Max'] = pd.to_numeric(df['Verified (ft)'])
    df = df.resample('D').max()
    return df
```

```python
def run_sarima(df, col):

    # Iterate through all possible differences.
    for d in range(1, len(df)):

        # Print out a counter and the corresponding p-value.
        print(f'Checking difference of {d}.')
        print(f'p-value = {interpret_dftest(adfuller(df[col].diff(12).diff(d).dropna()))["p-value"]}.')

        # If our data, differenced by d time periods, are stationary, print that out!
        if interpret_dftest(adfuller(df[col].diff(12).diff(d).dropna()))['p-value'] < 0.05:
            print(f'Differencing our time series by d={d} yields a stationary time series!')
            break

        print()

    model = SARIMAX(endog = df[col].dropna(), # endog = Y variable
                  order = (0,1,1), seasonal_order=(0, 1, 1, 12), enforce_invertibility=False) # (p, d, q) <-- talk about q later!

    model_fit = model.fit()
    y_hat = model_fit.predict()
              
    plt.figure() 
    plt.title(col)
    plt.plot(y_hat);
          
          
```
